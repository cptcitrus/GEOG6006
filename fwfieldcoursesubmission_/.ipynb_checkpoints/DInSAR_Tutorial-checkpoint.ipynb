# Data Management and Preprocessing

Here, we are going to read in our Sentinel-1 Interferometric Wide data and create all the necessary parameter files required for GAMMA. 

In this analysis, we are going to use two-pass differential interferometry. This requires two images, one before the deformation event and and one after.

Required Data:

- Two Sentinel-1 IW SLC images. Images can be downloaded from Alaska Satellite Facility or Copernicus Hub. The folders containing the data must be unzipped and located in the working directory.
- SRTM (or other DEM) that completely covers the area of the image you plan to use as your reference image (typically the image captured earlier in time). This must be in .tif format, and heights must be related to WGS 84

Within the working directory you need three folders:

- slc_TOPS (empty at the start of processing)
- slc_TOPS_deramp (empty at the start of processing)
- DEM (contains your DEM file)

Note that if you choose to follow this tutorial with different imagery, every time there is a date ("ex. 20210503"), it must be changed to match your image.

# Coregistration
Now it is time to coregister our two images. Very simply, we want to precisely align them at the sub-pixel level, which is necessscary for InSAR. The following steps will walk you through manual coregistration. Alternatively, there is a script created by GAMMA called ScanSAR_coreg.py to handle the majority of the coregistration procedure. As an input, it requires two SLCs and the height map we created during the earlier geocoding process.


Here, we will be applying mor than one method applied teratively improve the co-registration refinement. First, a matching procedure using  offset_pwr and then a spectral diversity method (script S1_coreg_overlap) which considers the interferometric phase of the burst overlap region. # Estimate coherence

In this section, we are going to create and export a coherence image, which is the phase correlation between are two images. Coherence is useful data on its own, as it is an indicator of image quality, and can be used to mask problem pixels during the phase unwrapping process.

# Phase Unwrapping

Since the phase is only known to modulo 2Pi in a wrapped interferogram, it must be unwrapped before it can be converted to geophysical units. Here, we are going to use the minimum cost flow (MCF) phase unwrapping method. MCF treats the phase unwrapping problem as a flow network, where each pixel is a node, and the phase differences between them are paths (or edges) in the network. The MCF algorithm assigns "costs" to these paths based on how much the phase changes between pixels. It then calculates the minimum cost flow through the network that results in a continuous unwrapped phase. In simple terms, the MCF algorithm tries to connect the phase inconsistencies by finding the cheapest way to connect the residues (Eineder et al., 1998).

```python
#Make a validity mask out of the coherence layer. The coherence threshold can be adjusted for each pair depending on how much phase noise there is.
#A threshold of .6 means that only pixels with coherence >= .6 will be unwrapped. I recommend starting with a threshold around here, looking at the results,
#and then adjusting.

pg.rascc_mask('20180916_Out_20180928_Out.diff.cc', '-', 2574, 1, 1, 0, 1, 1, .6)
```

```python
#Here we are phase unwrapping using the minimum cost flow algorithm.

pg.mcf('20180916_Out_20180928_Out.diff','20180916_Out_20180928_Out.diff.cc','20180916_Out_20180928_Out.diff.cc_mask.bmp', '20180916_Out_20180928_Out.diff.unw', '2574', '-', '-', '-', '-', '-', '-', '-', '-', '-', '-', 0 )
```

```python
#Converting the unwrapped interferogram in line of site displacement to vertical displacement in (m).

pg.dispmap("20180916_Out_20180928_Out.diff.unw", '20180916.hgt', "20180916.slc.par", "20180916_Out_20180928_Out.off", "20180916_Out_20180928_Out.diff.unw.displ", 1)
```
